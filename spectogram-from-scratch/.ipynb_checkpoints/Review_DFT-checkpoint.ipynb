```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(1,10,100)
f = 1/10
y = np.exp(1j*6.18*f*t[0]) #2*pi = 6.18
y_end = np.exp(1j*6.18*f*t[-1])

fig, ax = plt.subplots(1,1, figsize=(10,5))
ax.scatter(y.real,y.imag)
ax.scatter(y_end.real,y_end.imag,c='orange')
ax.set_xlim(-2,2)
ax.set_ylim(-1,1)
```

```python
plt.plot(t,y.imag)
plt.plot(t,y.real)
```

```python
original_ts = [0, 0.707, 1, 0.707, 0, -0.707, -1, -0.707]
L = len(original_ts)
plt.plot(original_ts,"p")
plt.title("original time series of length L={}".format(L))
plt.show()
```

```python
def get_xn(Xs,n):
    '''
    calculate the Fourier coefficient X_n of 
    Discrete Fourier Transform (DFT)
    '''
    L  = len(Xs)
    ks = np.arange(0,L,1)
    xn = np.sum(Xs*np.exp((1j*2*np.pi*ks*n)/L))/L
    return(xn)
```

```python
original_ts
```

```python
xs =[]
for n in range(L):
    xn = get_xn(original_ts,n)
    xs.append(xn)
    print("Fourier coefficient: x_{}={:+3.2f}".format(n,xn))
```
