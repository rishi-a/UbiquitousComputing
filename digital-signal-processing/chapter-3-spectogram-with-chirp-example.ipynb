# Spectrogram
Our objective is to visualise the frequency component of chirp at different period of time. This is achieved using Short-Term Fourier Transform. At first, we will use existing Python libraries to quickly plot the STFT or Spectrogram of the chirp signal. After that, we will write our own code to visualise spectrogram.<br><br>

**Recall**<br>
Spectrum showed us the relationship between Frequency and Amplitude. Spetrogram or STFt will show the rleationship between Spectrum, Amplitude and Time. It will tell us, what Frequency component is present with what Amplitude at what time?


```python
import numpy as np
import math
import matplotlib.pyplot as plt
from IPython.display import Audio
```

### Lets recall the spectrum of chirp from our last chapter

```python
def spectrum(signal, sr):
    A = np.fft.fft(signal, norm="backward")
    F = np.fft.fftfreq(signal.shape[0], d=1/sr)
    A = np.abs(A)
    return F[0:sr//2], A[0:sr//2]
```

```python
# linear chirp
sr = 16000
t = np.linspace(0,5,sr*5)
f = np.linspace(220,880,len(t))
dt = np.diff(t, prepend=0)
dphi = 2*np.pi*f*dt
phases = np.cumsum(dphi)
phases = np.insert(phases, 0, 0)
x1 = np.sin(phases)
F,A = spectrum(x1, sr)
plt.plot(F,A)
plt.xlim(0,1200)
audio = Audio(data=x1, rate=sr)
audio
```

```python
sr = 16000
start, end = np.log10(220), np.log10(880)
t = np.linspace(0,5,sr*5)
f_l = np.logspace(start, end, len(t))
dt = np.diff(t, prepend=0)
dphi = 2*np.pi*f_l*dt
phases = np.cumsum(dphi)
phases = np.insert(phases, 0, 0)
x2 = np.sin(phases)
F,A = spectrum(x2, sr)
plt.plot(F,A)
plt.xlim(0,1200)
audio = Audio(data=x2, rate=sr)
audio
```

```python
import librosa
import librosa.display
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,4))
n_fft = 1024
hop_length = n_fft//4 #the default value in librosa 
S = np.abs(librosa.stft(x1,hop_length=hop_length,n_fft=n_fft))
img = librosa.display.specshow(librosa.amplitude_to_db(S,ref=np.max),y_axis='linear', ax = ax, x_axis='time', sr=sr)
#ax.set_ylim(50,1500)
```
