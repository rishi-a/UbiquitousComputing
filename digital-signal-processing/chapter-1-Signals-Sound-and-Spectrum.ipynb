# Introduction to Signal
Signals are quantity that varies with time. Example, a sound signal or a video.
The simplest sound I can think of is that of sine wave. In high school, learned that a sine wave is represented by,

$$x = A\sin(2\pi ft)$$


```python
import numpy as np
import math
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
sr = 1000
f = 3
t = np.linspace(0,1,sr)
x = np.sin(2*np.pi*f*t)
plt.plot(t,x);
```

```python
audio = Audio(data=x, rate=sr)
audio
```

## What are Periods are Cycles ?
The signal that we saw above is a **periodic signal**. Periodic signals repeat themselves after some period of time. These repetations are known as **cycles**. The duration of each cycle is called a **period**

## What is Frequency?
The frequency of a signal is the number of cycles per second.

#### Can you now tell how period and frequency are related?

## What is Phase ?
It is the number of cycles since start time expressed in radians. Each cycle is $2\pi$ radians

## Intermixing frequenies and Timbre

```python
def generateSignal(t,sr,f, phi):
    t = np.linspace(0,t,sr*t)
    return t,np.sin(2*np.pi*f*t+phi);
```

```python
t, x1 = generateSignal(1,1000,440, 0)
audio = Audio(data=x1, rate=1000)
audio
```

```python
# Timbre
# A Piano playing A4 (440 Hz): https://www.youtube.com/watch?v=C__Ug0K4sdU
# A violin playing A4 (440 Hz): https://www.youtube.com/watch?v=j0FynYzQvcM
```

```python
# Intermixing frequencies
t, x1 = generateSignal(1,1000,440, 0)
t, x2 = generateSignal(1,1000,100, np.pi)


fig, ax = plt.subplots(2,1, figsize=(14,6))
ax[0].plot(t,x1)
ax[1].plot(t,x2)
```

```python
audio = Audio(data=x1, rate=1000)
audio
```

```python
# intermix these signals.
mix_x = x1+x2
fig, ax = plt.subplots(1,1, figsize=(14,6))
ax.plot(t,mix_x)
audio = Audio(data=mix_x, rate=1000)
audio
```

## Spectral Decomposition of Signals

```python
signal = x1
A = np.fft.fft(signal, norm="backward")
F = np.fft.fftfreq(signal.shape[0], d=1/1000)
A = np.abs(A)

plt.plot(F[0:sr//2],A[0:sr//2])
```

```python
signal = x2
A = np.fft.fft(signal, norm="backward")
F = np.fft.fftfreq(signal.shape[0], d=1/1000)
A = np.abs(A)

plt.plot(F[0:sr//2],A[0:sr//2])
```

```python
signal = mix_x
A = np.fft.fft(signal, norm="backward")
F = np.fft.fftfreq(signal.shape[0], d=1/1000)
A = np.abs(A)

plt.plot(F[0:sr//2],A[0:sr//2])
```
