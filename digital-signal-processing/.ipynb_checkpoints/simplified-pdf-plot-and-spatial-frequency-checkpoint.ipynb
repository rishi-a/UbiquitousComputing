```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Generate some example data
np.random.seed(0)
data = np.concatenate([np.random.normal(-2, 1, 500), np.random.normal(2, 0.5, 300)])

# Define the kernel function (Gaussian kernel)
def gaussian_kernel(x):
    return np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)

# Set the bandwidth parameter
bandwidth = 0.5

# Generate a range of x values to evaluate the estimated PDF
x = np.linspace(min(data) - 2, max(data) + 2, 100)

# Estimate the PDF by summing up the contributions from each data point
estimated_pdf = np.zeros_like(x)
for xi in data:
    estimated_pdf += gaussian_kernel((x - xi) / bandwidth)
print(estimated_pdf)

# Normalize the estimated PDF
estimated_pdf /= (len(data) * bandwidth)

# Plot the observed data and the estimated PDF
plt.hist(data, bins=30, density=True, alpha=0.5, label='Observed Data')
plt.plot(x, estimated_pdf, 'r-', label='Estimated PDF')
plt.xlabel('X')
plt.ylabel('Density')
plt.title('Kernel Density Estimation')
plt.legend()
plt.show()
```
