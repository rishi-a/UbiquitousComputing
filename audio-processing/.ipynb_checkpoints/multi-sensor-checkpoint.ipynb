```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import scipy
```

```python
data11 = pd.read_csv('data/21-12-Idle-COM11.csv', sep=' ', names=['Date', 'Time', 'Ax', 'Ay', 'Az','Gx', 'Gy', 'Gz','Mx', 'My', 'Mz','Proximity','Gesture','Temperature','Humidity','Preassure'])
data11.Time = pd.DatetimeIndex(data11.Time)
data7 = pd.read_csv('data/21-12-Idle-COM7.csv', sep=' ', names=['Date', 'Time', 'Ax', 'Ay', 'Az','Gx', 'Gy', 'Gz','Mx', 'My', 'Mz','Proximity','Gesture','Temperature','Humidity','Preassure'])
data7.Time = pd.DatetimeIndex(data7.Time)
```

```python
data11.index = data11.Time
data11 = data11.drop(['Date', 'Time'], axis=1)
data7.index = data7.Time
data7 = data7.drop(['Date', 'Time'], axis=1)
```

```python
#combine preassure data
pressure = pd.merge(data11['Preassure'], data7['Preassure'], on='Time')
#pressure.index = pressure.Time
pressure.head()
```

```python
pressure.plot(figsize=(10,7), subplots=True, title="Preassure Sensor 1 and 2")
```

```python
data13 = pd.read_csv('data/22-12-Idle-COM13.csv', sep=' ', names=['Date', 'Time', 'Ax', 'Ay', 'Az','Gx', 'Gy', 'Gz','Mx', 'My', 'Mz','Proximity','Gesture','Temperature','Humidity','Preassure'])
data13.Time = pd.DatetimeIndex(data13.Time)
data13.index = data13.Time
data13 = data13.drop(['Date', 'Time'], axis=1)
data13['Preassure'].plot(figsize=(10,3.5), title="Preassure Sensor 3")
```

```python
#combine preassure data
humidity = pd.merge(data11['Humidity'], data7['Humidity'], on='Time')
#pressure.index = pressure.Time
humidity.head()
humidity.plot(figsize=(10,5), subplots=True, title="Humidity Sensor 1 and 2")
```

```python
data13['Humidity'].plot(figsize=(10,3.5), title="Humidity Sensor 3")
```

```python
#combine preassure data
temp = pd.merge(data11['Temperature'], data7['Temperature'], on='Time')
#pressure.index = pressure.Time
temp.head()
temp.plot(figsize=(10,5), subplots=True, title="Temperature Sensor 1 and 2")
```

```python
data13['Temperature'].plot(figsize=(10,3.5), title="Temperature Sensor 3")
```
