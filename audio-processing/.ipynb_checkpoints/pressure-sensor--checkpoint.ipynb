This notebook demonstrates the use of pressure sensor for calculating the Flow vs Volume curve, typically obtained using spirometer subjected to the monitoring of FVC Manoeuvre. 

The sensor used is [LPS22HB](https://www.st.com/en/mems-and-sensors/lps22hb.html) found in the Arduino Nano 33 BLE Sense. The sensor has an atmospheric pressure range of $26 KPa \leq p \leq126 KPa$. Current Consumption $3 \mu A$. For this experiment, the sampling rate was 10 Hz.

A unit of atmospheric preasssure = 101,325 Pa = 101.325 KPa = 1013.25 HPa = 760mm of Hg<br>


### Experiment:
**Aim:** To obtain Flow vs Volume curve using preassure sensor.<br>
**Approach**: We used two preassure sensor. One preassure sensor was fixed inside an N95 mask. The other one was idle on a bench top. We used two preassure sensor to differentiate the change in preassure inside the mask with that of the atmospheric preassure.  

```python
#load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import scipy
```

```python
# 10hz data
#read the data from the first pressure sensor which is idle on the table
data11 = pd.read_csv('data/e5-com11.csv', sep=' ', names=['Date', 'Time', 'Preassure 1'])
data11.Time = pd.DatetimeIndex(data11.Time)
#read the data fromt the second pressure sensor which is attached to the mask
data13 = pd.read_csv('data/e5-com13.csv', sep=' ', names=['Date', 'Time', 'Preassure 2'])
data13.Time = pd.DatetimeIndex(data13.Time)

#merge both the data so that we can compare
combined1113 = pd.merge(data11, data13, on='Time')
combined1113.index = combined1113.Time
combined1113 = combined1113.drop(['Date_x', 'Date_y', 'Time'], axis=1)
combined1113.head()
```

```python
# there is a average difference of 0.05kPa between both the pressure sensor when they are idle. 
# so we compensate for this difference
combined1113['Preassure 2'] = combined1113['Preassure 2'] - np.mean(combined1113['Preassure 2'] - combined1113['Preassure 1'])
```

```python
ax = combined1113.plot(figsize=(20,5))
ax.set_xlabel('Timestamp')
ax.set_ylabel('Preassure')
ax.legend(['Sensor 1 (On benchtop)', 'Sensor 2 (placed inside the mask worn by a invetigator)'])
```

The peak in the orange plot (at Timestamp 10:34:00) shows an FVC manoeuvre. I performed only one FVC manoeuvre during the entire timeline of this experiment. The relation between Flow and Preassure is related as $Flow \propto \sqrt{Preassure}$, Thus

```python
#square root
combined1113['Preassure 1'] = (combined1113['Preassure 1'])**(1/2)
combined1113['Preassure 2'] = (combined1113['Preassure 2'])**(1/2)
ax = combined1113.plot(figsize=(20,5))
ax.set_xlabel('Timestamp')
ax.set_ylabel('Preassure')
ax.legend(['Sensor 1 (On benchtop)', 'Sensor 2 (placed inside the mask worn by a invetigator)'])
```

Our area of interest in the curve is the part of exhalation during the FVC manoeuvre. Thus, we extract that portion from the signal

```python
#area of interest
df = combined1113['2020-12-17 10:34:2.0000':'2020-12-17 10:34:6.00000']['Preassure 2']
```

```python
ax = df.plot(figsize=(8,5))
ax.set_xlabel('Timestamp')
ax.set_ylabel('Proportionate Flow')
ax.legend(['Sensor 2 (placed inside the mask worn by a invetigator)'])
```

The make the FVC curve; we need volume in as an independent variable on the x-axis. We estimate volume as a cumulative sum of flow.

```python
df1 = df.cumsum(axis = 0)
ax = df1.plot()
ax.set_xlabel('Timestamp')
ax.set_ylabel('Proportionate Volume')
```

Now, we combine our estimated flow and volume in a single plot,

```python
newdf = pd.DataFrame()
newdf['Flow'] = df
newdf['Volume'] = df1
ax = newdf.plot(y='Flow', x='Volume')
ax.set_xlabel('Volume')
ax.set_ylabel('Flow')
```
