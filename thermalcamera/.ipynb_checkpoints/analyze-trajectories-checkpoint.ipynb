```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from scipy import spatial
import cv2
import scipy as sp
from scipy import signal
```

```python
def click_event(event, x, y, flags, params):
    global old_points, first_point;
    if event == cv2.EVENT_LBUTTONDOWN:
        old_points = np.array([[x,y,]], dtype=np.float32)
        first_point = old_points
# lucas kanade params
lk_param = dict(winSize=(350,350), maxLevel=24, criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10,0.03))
    
```

```python
def compute_fft(s, sampling_rate, n = None, scale_amplitudes = True):
    '''Computes an FFT on signal s using numpy.fft.fft.
    
       Parameters:
        s (np.array): the signal
        sampling_rate (num): sampling rate
        n (integer): If n is smaller than the length of the input, the input is cropped. If n is 
            larger, the input is padded with zeros. If n is not given, the length of the input signal 
            is used (i.e., len(s))
        scale_amplitudes (boolean): If true, the spectrum amplitudes are scaled by 2/len(s)
    '''
    if n == None:
        n = len(s)
        
    fft_result = np.fft.fft(s, n)
    num_freq_bins = len(fft_result)
    fft_freqs = np.fft.fftfreq(num_freq_bins, d = 1 / sampling_rate)
    half_freq_bins = num_freq_bins // 2
 
    fft_freqs = fft_freqs[:half_freq_bins]
    fft_result = fft_result[:half_freq_bins]
    fft_amplitudes = np.abs(fft_result)
    
    if scale_amplitudes is True:
        fft_amplitudes = 2 * fft_amplitudes / (len(s))
    
    return (fft_freqs, fft_amplitudes)

def maskCircularArea(gray_frame, x_loc, y_loc, circleRadius):
    '''
    This functions accepts the following param:
        gray_frame: A frame of a opencv video
        x_loc: The X Coordinate of a circle
        y_loc: The Y coordinate of a circle
        circleRadius: The radius of the circle to be created
    
    This functions returns sum of all the pixels covered by the circle over the frame. 
    Higher the radius of the circle, greater the returned sum
    '''
    cv2.circle(frame, (int(x_loc),int(y_loc)),circleRadius,(255,0,0),-1)
    # create the circle with new point over the mask as well
    circle_mask = cv2.circle(mask, (int(x_loc), int(y_loc)),circleRadius,(255,0,0),-1)
    masked_data = cv2.bitwise_and(gray_frame,gray_frame,mask=circle_mask)
    tom, thres = cv2.threshold(mask, 1,255, cv2.THRESH_BINARY)
    contours = cv2.findContours(thres, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    x,y,w,h = cv2.boundingRect(contours[0][0])
    crop = masked_data[y:y+h,x:x+w]

    # now that we have the pixels corresponding to the ROI, lets sum of all pixels.
    sumPixels = 255-np.average(crop[:,:])
    return sumPixels
```

```python
for nearestNegh in range(1,3,1):
    
    print("Processing for Neighbor {}".format(nearestNegh))
    
    # control vars
    showDifference = False
    isSparseOF = False
    pt = [357, 176]
    
    # load the data
    x = torch.load('../../pips/savedTraje/P11/2.pt').cpu()
    x = np.rint(x.numpy()).astype(int)
    
    # init two arrays
    deltaChange = np.zeros((x.shape[1]-1,2))
    pixelLocChange = np.zeros((x.shape[1],2))
    # a dict where we will store the sum of pixels of a frame. Spread across all frames
    if nearestNegh == 1: 
        mixedPixelValues = {i:0 for i in range(0,x.shape[1],1)}
    
    # update the arrays for each frame in vide
    for i in range(0,x.shape[1],8):
        firstFrame = x[0,i,:,:]
        indexToTrack = spatial.KDTree(firstFrame).query(pt, k=[nearestNegh])[1]
        firstPoint = firstFrame[indexToTrack] # <-- the nearest point 
        deltaChange[i:i+7] = np.diff(x[0,i:i+8,indexToTrack[0],:],axis=0)
        pixelLocChange[i:i+8] = x[0,i:i+8,indexToTrack[0],:]
    
    x_init = pt[0]
    y_init = pt[1]
    frameLeft = x.shape[1] #set this later
    fps = 8.737238044062332
    
    filenameWithPath = 'G:/Other computers/My MacBook Air/thermal-project-data/cycling/20221014T110944.MP4'
    cap = cv2.VideoCapture(filenameWithPath)
    # get the frame rate
    fps = cap.get(cv2.CAP_PROP_FPS)
    correct,frame = cap.read()
    height, width, channel = frame.shape

    print("Original Height = {}, Original Width = {}".format(height, width))
    frame = cv2.resize(frame, (640+50, 360+50))
    # convert the first frame to grayscale
    prev_gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    # keep a tab of the frame count
    frameNumber = 1
    frameLeft-=1
    print("Total number of frames = {}".format(frameLeft))

    intensityMap = []
    # list to store all the ROI's sum of pixels for each frame of each video
    rrSignal = {}
    dropped=0
    
    if isSparseOF:
        # show the first frame. imshow requires param to be in BGR format unlike Matplotlib's RGB
        cv2.imshow('image', frame)
        #b,g,r = cv2.split(frame)
        #frame = cv2.merge((r,g,b))
        # hook the function to be called in case of mouse click
        cv2.setMouseCallback('image', click_event)
        # wait infinitely until a key is pressed.
        cv2.waitKey(0)
        
    
    #at this point we have already read one frame
    while frameLeft and correct:
        # read the first frame of the video
        correct,frame = cap.read()
        frame = cv2.resize(frame, (640+50, 360+50))
        gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

        # get the size of the frame
        height, width, channel = frame.shape
        # create a all-zeros mask, we are doing this to get all the pixels covered by the circle
        mask = np.zeros((height, width), np.uint8)
        if isSparseOF:
            # get the new points for optical flow
            new_points, status, error = cv2.calcOpticalFlowPyrLK(prev_gray, gray_frame, old_points, None, **lk_param)
            # update the old frame to the current frame
            prev_gray = gray_frame.copy()
            # update the old points to new points
            old_points = new_points
            # make arrangements to plot the points
            x,y = new_points.ravel()
            x_i, y_i = first_point.ravel()
            cv2.circle(frame, (int(x),int(y)),2,(255,0,255),-1)
            if frameNumber==2: print("Initial X {}, Y {}".format(x_i,y_i))
         # store or show pixel intensity
        try:
            if showDifference:

                if frameNumber%8==0:
                    x_init = int(pixelLocChange[frameNumber][0])
                    y_init = int(pixelLocChange[frameNumber][1])
                    x_loc = x_init
                    y_loc = y_init
                else:
                    x_loc = x_init+int(deltaChange[frameNumber-1][0])
                    y_loc = y_init+int(deltaChange[frameNumber-1][1])
                    x_init = x_loc
                    y_init = y_loc

                #print("FN = {}, XY = {},{}".format(frameNumber,x_loc,y_loc))
                intensityMap.append(gray_frame[x_loc][y_loc])
                # for each frame, store the sumPixel in an array. This array is the signal for RR
                rrSignal[frameNumber] = maskCircularArea(gray_frame, x_loc, y_loc, 100)
            else:
                x_loc = int(pixelLocChange[frameNumber][0])
                y_loc = int(pixelLocChange[frameNumber][1])
                intensityMap.append(gray_frame[x_loc][y_loc])
                rrSignal[frameNumber] = maskCircularArea(gray_frame, x_loc, y_loc, 100)
        except:
            dropped+=1
            #print(x_loc, y_loc)
        cv2.imshow('image', frame)
        key = cv2.waitKey(int((1/fps)*100))
        if key == 27 or key == ord('q'):
            break
        if frameLeft%100==0: print("{} Frames Left To Be Processed".format(frameLeft))
        frameLeft-=1;
        frameNumber+=1
    print("Percentage of frames dropped = {}\n\n\n".format((dropped/frameNumber)*100))
    for k in rrSignal.keys():
        mixedPixelValues[k]+=rrSignal[k]
cap.release()
cv2.destroyAllWindows()
```

```python
signal = list(mixedPixelValues.values())
size = 5
window = sp.signal.gaussian(M=size, std=5)
window /= window.sum()
filteredA = np.convolve(signal[1:], window, mode='same')



fig, ax = plt.subplots(3,1, figsize=(17,8))
F,A = compute_fft(filteredA, fps)
highestFreq = F[np.argmax(A[1:])+1]
print(highestFreq)
ax[0].plot([i/fps for i in range(0,len(signal[1:]))], signal[1:])
ax[1].plot([i/fps for i in range(0,len(signal[1:]))], filteredA)

ax[2].plot(F[1:],A[1:])
#ax[1].set_ylim(12000,15000)




```
