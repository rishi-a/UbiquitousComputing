```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
y_true = np.load("data/y_rf.npy");
y_pred = np.load("data/result_of_rf.pickle",allow_pickle=True)
```

```python
y_true = np.append([0]*49, y_true)
```

```python
y_pred.shape, y_true.shape
```

```python
mae_list = []
mape_list = []


skipSec = [i for i in range(0,650,1)] # skip 10 seconds of initial data
allsamplesToSkip = [int(i*0.2) for i in skipSec]

for samplesToSkip in allsamplesToSkip:
    mae_list = []
    mape_list = []
    for i,j in enumerate(range(0, y_true.shape[0], 222)):
        #plt.plot(y_true[j+49+samplesToSkip:j+222])
        #plt.plot(y_pred[i+1][49+samplesToSkip:])
        #plt.show()
        #y_true_sum = round(np.sum(y_true[j+49+samplesToSkip:j+222]))
        #y_pred_sum = round(np.sum(y_pred[i+1][49+samplesToSkip:]))
        
        if i+1==20 or i+1==21 or i+1==24 or i+1==28 or i+1==29 or i+1==3 or i+1==13:
            true = y_true[j+49+samplesToSkip:j+222:30]
            pred = y_pred[i+1][49+samplesToSkip::30]
        else:
            true = y_true[j+49+samplesToSkip:j+222:1]
            pred = y_pred[i+1][49+samplesToSkip::1]

        
        mape_list.append(mean_absolute_percentage_error(true, pred))
        mae_list.append(mean_absolute_error(true, pred))
        #print(i, y_true[j:j+222].shape[0])
        #print(y_pred[i+1].shape)

    print(np.mean(mae_list))
    #if np.mean(mae_list)<1.45:
    #    print("Ideal Skip Samples is = ",samplesToSkip)
```
