```python
import numpy as np
import scipy as sp
from scipy import signal
import matplotlib.pyplot as plot
```

```python
def compute_fft(s, sampling_rate, n = None, scale_amplitudes = True):
    '''Computes an FFT on signal s using numpy.fft.fft.
    
       Parameters:
        s (np.array): the signal
        sampling_rate (num): sampling rate
        n (integer): If n is smaller than the length of the input, the input is cropped. If n is 
            larger, the input is padded with zeros. If n is not given, the length of the input signal 
            is used (i.e., len(s))
        scale_amplitudes (boolean): If true, the spectrum amplitudes are scaled by 2/len(s)
    '''
    if n == None:
        n = len(s)
        
    fft_result = np.fft.fft(s, n)
    num_freq_bins = len(fft_result)
    fft_freqs = np.fft.fftfreq(num_freq_bins, d = 1 / sampling_rate)
    half_freq_bins = num_freq_bins // 2
 
    fft_freqs = fft_freqs[:half_freq_bins]
    fft_result = fft_result[:half_freq_bins]
    fft_amplitudes = np.abs(fft_result)
    
    if scale_amplitudes is True:
        fft_amplitudes = 2 * fft_amplitudes / (len(s))
    
    return (fft_freqs, fft_amplitudes)
```

```python
t = np.linspace(0, 1, 44100, endpoint=True)
squareSignal = signal.square(2 * np.pi * 440 * t)
plot.plot(t, squareSignal)
plot.xlim((0,0.010))
```

```python
F_square, A_square = compute_fft(squareSignal, 44100,scale_amplitudes=False)
window = np.ones(11)
window /= sum(window)
convolved_square = np.convolve(squareSignal, window, mode='valid')
F_square_c, A_square_c = compute_fft(convolved_square, 44100,scale_amplitudes=False)
plot.plot(F_square,A_square)
plot.plot(F_square_c, A_square_c)
plot.legend(['Before Smoothing','After Smoothing'])
#plot.xlim((0,5000))
```

```python
ratio = A_square_c/A_square[:-5]
ratio[A_square[:-5]<560] = 0
plot.plot(F_square_c,ratio)
plot.legend(['Amplitude Ratio'])
plot.ylim(0,0.10)
```

### Why does the ratio bounce for lower amplitude ratio? 
Thats becuase, convolution operation means multiplication of frequency domain signals.

```python
window_appended = np.append(window,np.zeros(len(squareSignal)))
F_window, A_window = compute_fft(window_appended, 44100,scale_amplitudes=False)
plot.plot(F_window, A_window)
```

As seen in the plot above, the DFT of the window has the _bouncing effect_ which appeared in our filtered square wave signal above. Technically speaking, the DFT of the window is actually the DFT of the window which is padded with zeros for the window to be of same length as the original filtered signal above.

**The moving average window of ones that we used above is not a very good low pass filter. As we saw above, that filter behaves as a square wave. We just saw that the spectrum of square wave has _sidelobes_. Sidelobes are nothing but high frequency harmonics that drop of proportionally to $\frac{1}{f}$. Apparently, that fall off is slow to filter signals.**
